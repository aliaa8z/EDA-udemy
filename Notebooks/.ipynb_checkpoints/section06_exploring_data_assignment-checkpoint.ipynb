## Read in the data

```python
# import the data 
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv('../Data/happiness_survey_data.csv')
df.head()
```

```python
# there are 2089 rows and 6 columns of data
df.shape
```

```python
# the years range from 2005 - 2021
# happiness scores range from 2 - 8
df.describe()
```

## Explore the data

```python
# Filter out any data before 2010 and after 2019
df_filter = df[(df.year < 2010) | (df.year > 2019)]
df_filter.head()
```

```python
# Group the data by country and calculate the maximum happiness score for each one
df_gr = df_filter.groupby('country_name')['happiness_score'].max().to_frame()
df_gr
```

```python
# Sort the grouped countries by happiness score and return the top five
df_sort = df_gr.sort_values(by = 'happiness_score', ascending= False)[:5]
df_sort
```

```python
# Group the data by country and calculate the average happiness score for each one
df_mean = df_filter.groupby('country_name')['happiness_score'].mean().to_frame()
df_mean
```

```python
# Sort the grouped countries by happiness score and return the top five
df_filter.groupby('country_name')['happiness_score'].mean().sort_values(ascending= False).to_frame()
```
